```python
# Criando uma Variavel X para exemplificação
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
# função numpy "linspace" para criar um intervalo de valores

# Obtendo Cosseno e Seno de X para exemplificação
c, s = np.cos(x), np.sin(x)
```

```python
# Gerando um grafico simples com o matplotlib selecionando os valores

# Plotando x com cosseno
plt.plot(x, c, color='blue', linewidth=1.0, linestyle = '--')

# Plotando x com seno
plt.plot(x, s, color='green', linewidth=1.0, linestyle = '-')

# Inserindo um Titulo ao grafico
plt.title('Grafico de Exemplo 01')

#  Limite do eixo X para escala de valores
plt.xlim(-4.0, 4.0)

# Escala de Valores no eixo x
plt.xticks(np.linspace(-4,4,9, endpoint=True))

# Nomendo o Eixo X
plt.xlabel('Eixo X')

# Limite do eixo Y para a escala de valores
plt.ylim(-1.0,1.0)

# escala de Valores no eixo y
plt.yticks(np.linspace(-1,1,5, endpoint=True))

# Nomendo o eixo y
plt.ylabel('Eixo Y')

# Visualizar o Grafico
plt.show()
```

```python

plt.figure(figsize=(8,5), dpi=80)
ax = plt.subplot(111)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-",  label="sine")

# Inserindo um Titulo ao grafico
plt.title('Grafico de Exemplo 05')

plt.xlim(X.min()*1.1, X.max()*1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.ylim(C.min()*1.1,C.max()*1.1)
plt.yticks([-1, +1],
           [r'$-1$', r'$+1$'])

t = 2*np.pi/3
plt.plot([t,t],[0,np.cos(t)],
         color ='blue',  linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.cos(t),], 50, color ='blue')
plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)),  xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t,t],[0,np.sin(t)],
         color ='red',  linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.sin(t),], 50, color ='red')
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)),  xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.legend(loc='upper left', frameon=False)
#plt.savefig("figura.png",dpi=72)
plt.show()
```

##  ``pyplot.boxplot`` matplotlib

* Desenhe um gráfico de caixa e bigode.

# Boxplot

Usado para visualizar a distribuição dos dados.

## Interpretação do Boxplot

image.png

* Q1: Primeiro quartil. 25% dos pontos são menores que o Q1 e 75% dos pontos são maiores.
* Q2: Mediana
* Q3: Terceiro quartil, 75% dos pontos são menores e 25% são maiores.
* IQR: Distancia Inter-Quartil, define o limite superior e inferior do plot. Limite superior: Q3 + 1.5*IQR
* Dados fora desse limite são chamados de **outliers**

Obs.: Existem diversas abordagens para definir o tamanho do boxplot.

image.png


```python
# Gerando dados para exemplificação
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))

# Criando uma pltagem de um grafico de caixa
plt.boxplot(data)
plt.show()
```

## ``pyplot.contourf`` matplotlib

* Trace contornos preenchidos

```python
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X, Y, f(X,Y), 10, alpha=0.99, cmap='jet')
#C = plt.contour(X, Y, f(X,Y), 8, colors='black' , linewidth=.5)
plt.show()
```

## ``pyplot.imshow`` matplotlib

* Exiba dados como uma imagem, ou seja, em um raster 2D regular. A entrada pode ser dados RGB(A) reais ou dados escalares 2D, que será renderizado como uma imagem pseudocolorida

```python
# Criando uma função python para exemplificação
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

# Gerando dados para exemplificação
n = 10
x = np.linspace(-3,3,4*n)
y = np.linspace(-3,3,3*n)
X,Y = np.meshgrid(x,y)
```

```python
# Criando uma plotagem de uma imgem colorida
plt.imshow(f(X,Y))
plt.show()
```

```python
plt.subplot(311)
plt.imshow(np.random.random((100, 100)))
plt.subplot(321)
plt.imshow(np.random.random((100, 100)))

plt.subplots_adjust(bottom=0.1, right=0.8, top=0.9)
cax = plt.axes([0.85, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)

plt.show()
```

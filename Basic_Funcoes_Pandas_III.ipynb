## <center> Carregando Dados em DataFrame Pandas </center>

```python
# Carregando Dados em DataFrame Pandas para Exemplificação
df_ex1 = pd.read_csv('pokemon_data.csv')

# Excluindo a Coluna "#"
df_ex1.drop('#', axis=1, inplace=True)
display(df_ex1.head())
```

## <center> Verificando Dados no DataFrame </center>

## ``copy`` DataFrame Pandas

* Faça uma cópia dos índices e dados deste objeto

```python
df_ex2 = df_ex1.copy()
```

## ``isnull`` Pandas

* Detecte valores ausentes nulos(Null / NaN)

```python
# Retorna um DataFrame onde, True para Nulos e False, para Não Nulos
display(df_ex1.isnull())
```

```python
# Detectando Valores nulos e aplicando a função soma para obter o total de elementos nulos
print(df_ex1.isnull().sum())
# Existem 386 valores nulos na coluna "Type 2"
```

## ``dropna`` DataFrame Pandas

* Excluir valores Nulos / Ausentes ( NaN, Null) linhas ou colunas

```python
# Verificando a forma do DataFrame antes de excluir valores nulos
print(f'Antes de Excluir: {df_ex1.shape}\n')

# Eliminar as Linhas com valores nulos/ausentes
df_ex1.dropna(inplace=True)

# Verificando a forma do DataFrame após a exclusão das linhas com valores nulos
print(f'Depois de Excluir: {df_ex1.shape}\n')

# Verificando a exclusão das linhas com valores nulos
print(df_ex1.isnull().sum())
```

```python
# Checando os indices após a exclusão das linhas com valores ausentes
df_ex1.index
# Os índices não se encontram em ordem após exclusão
```

## ``reset_index`` DataFrame Pandas

* Redefinindo o índice ou um nível do mesmo 

```python
# Reordenando o índice após exclusão
df_ex1.reset_index(drop=True, inplace=True)
display(df_ex1.head())
```

```python
# Checando o reordenamento do índice
df_ex1.index
# Reordenamento do índice Ok
```

## <center> Manipulando Dados Multi-Index </center>

## ``set_index`` DataFrame Pandas

* Defina o índice do ataFrame utilizando colunas existentes.

```python
# Craindo dataFrame para exeplificação
frame = pd.DataFrame(
    {'a':range(7), 
     'b':range(7, 0, -1),
     'c':['one', 'one', 'one', 'two', 'two', 'two', 'two'],
     'd':[0,1,2,0,1,2,3]
    })
display(frame)
```

##  ``unstack`` DataFrame Pandas

* Dinamize um nível dos rótulos de índice (necessariamente hierárquicos). Retorna um DataFrame com um novo nível de rótulos de coluna cujo nível mais interno consiste nos rótulos de índice dinâmico. Se o índice não for um MultiIndex, a saída será uma Série. 

```python
# Criando dados para exemplificação
data = pd.Series(np.random.randn(9), index=[[
    'a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],[
    1, 2, 3, 1, 3, 1, 2, 2, 3
]])

print(data.head(10))
```

```python
print(data.index)
```

```python
display(data.unstack())
```

## ``stack`` DataFrame Pandas

Empilhe os níveis prescritos das colunas ao índice. Retornar um DataFrame ou Série remodelado com multinível índice com um ou mais novos níveis mais internos em comparação com o atual Quadro de dados. Os novos níveis mais internos são criados girando o colunas do dataframe atual:

* se as colunas tiverem um único nível, a saída será uma Série;

* se as colunas tiverem vários níveis, o novo índice nível(s) é(são) retirado(s) do(s) nível(s) prescrito(s) e a saída é um DataFrame.

```python
print(data.unstack().stack())
```

## <center> Verificando Dados no DataFrame </center>

## ``nunique`` DataFrame Pandas

 * Contar o número de elementos distintos no eixo especificado.

## ``unique`` DataFrame Pandas

## ``duplicated`` DataFrame Pandas

* Retorna uma **série booleana denotando** linhas duplicadas.

```python
# Obtendo uma lista de valores booleanos para linhas duplicadas
df_ex1.duplicated().sum()
# Atrelando a função de soma para obter o total de linhas duplicadas
```

```python
# Checando dados redundantes, em duplicidade e somando tudo que existe
df_ex1.duplicated(subset = 'Type 1').sum()
# Existem 396 registros iguais, em duplicidade somente na coluna "Type 1" 
```

## ``drop_duplicates`` DataFrame Pandas

* Removendo linhas duplicadas retorna o DataFrame com as linhas duplicadas removidas.

```python
# Remover linhas com registros em redundancia (duplicidade)
df_ex1.drop_duplicates(keep='first', inplace=True)
# Removendo linhas em redundancia, em duplicidade, porém, mantendo o primeiro registro de cada duplicidade encontrada 
```

## ``replace`` DataFrame Pandas

* Substituir um valor / elemento por outro

```python
# Substituir  definitivamente, em todo DataFrame, os elementos especificados 
df_ex1.replace({True: 'Verdadeiro', False: 'Falso'}, inplace = True)
# Subtititui todo os elementos que forem "True", para "Verdadeiro" e todos os elementos que forem "False"  para "Falso"

display(df_ex1)
```

```python
# Chegar a substituição
print(df_ex1['Legendary'].value_counts())
```

# <center> Agregação e Agrupamento Pandas </center>

## ``groupby`` DataFrame Pandas

* Agrupe DataFrame usando um mapeador(map) ou por uma série de coluna

```python
print(data.groupby(level=0).sum())
```

```python
# Atribuindo a uma variavel o agrupamento especificado
df2 = df_ex1.groupby(['Name', 'Type 1', 'Type 2']).sum()
# Somando o agrupamento
display(df2.head())

# Reordenando o índice
df2 = df2.reset_index()

# Reorganizando o DataFrame de forma crescente de acordo com as colunas especificadas
df2 = df2[['Name', 'Type 1', 'Type 2', 'HP']].sort_values('Type 1')

display(df2.head(8))
```

```python
df_ex3 = df_ex2.drop(columns=['Name','Type 2'])
# Somar o agrupamento por "Type 1",
df_ex3.groupby('Type 1').sum()
```

## ``get_group`` DataFrame Pandas

*  Controi um DataFrame a apartir do groupby  

```python
display(df_ex2.groupby('Type 1').get_group('Bug'))
```

## ``agg`` DataFrame Pandas 

* Agregue usando uma ou mais operações no eixo especificado

```python
df_ex2.groupby('Type 1')['HP'].agg([min, np.median, max])
```

## ``aggregate`` DataFrame Pandas

* Agregue usando uma ou mais operações no eixo especificado

```python
df_ex2.groupby('Type 1')['HP'].aggregate([min, np.median, max])
```

```python
df_ex2.groupby('Type 1').agg({'HP':['max', 'min'],'Sp. Atk':[np.std, np.median]})['HP', 'max']
```

## ``apply`` DataFrame Pandas

* Aplique uma função ao longo de um eixo do DataFrame.

```python
# Criando uma função python para exemplificar 
def max_min(x):
    diferenca = x.max() - x.min()
    return diferenca

df_ex1.loc[:, ['HP', 'Sp. Atk']].apply(max_min)
```

```python
df_ex1[['HP', 'Sp. Atk']].apply(max_min)
```

## ``map`` DataFrame Pandas

* Aplicar uma função a um Dataframe elemento a elemento.

```python
# Criando uma função python para exemplificar 
def add_string(string):
    nova_string = 'Pokemon_'+string
    return nova_string

# Aplicando a função python, por meio do comendo do pandas, "map"
print(df_ex1['Name'].map(add_string))
```

```python
# Obtendo a raiz quadrada de cada elemento
df_ex1['HP'].map(np.sqrt)
```

## ``applymap`` DataFrame Pandas

* Aplicar uma função a um Dataframe elemento a elemento.

```python
# Obtendo a raiz quadrada de cada elemento
display(df_ex1.loc[:, ['HP', 'Sp. Atk', 'Attack', 'Defense']].applymap(np.sqrt))
```

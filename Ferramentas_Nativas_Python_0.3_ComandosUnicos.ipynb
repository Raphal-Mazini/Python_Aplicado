## ``Input``

O ``input()`` é uma função que permite com que se solicite uma informação ao usuário, para que ele possa preencher, seja para responder uma pergunta, seja para fazer o login em um sistema.

Esta função aceita todos tipos primitivos de dados, porém esta retorna sempre uma string, mas é possivel modificar o retorno com a inserção, junto ao ``input()`` a tipagem de interesse, entrentanto, ao iserir a tipagem de interesse, somante aceitará dados desta mesma tipagem de interesse, e tão somente este tipo poderá ser inserido 

```python
# Padrão, permite inserir qualquer tipo de dado, mas somente retorna string
nome = input('Insira seu nome: ')
print(f'A classe/tipo deste dado é : {type(nome)}')
print(nome)
```

```python
# Deste modo permite inserir somemente tipo de dado numerico, inteiro, e retorna um tipo inteiro 
numero = int(input('Insira um numero: '))
print(f'A classe/tipo deste dado é : {type(numero)}')
print(numero)
```

##  ``.isspace()``

Função ``.isspace()`` verifica se uma string contém apenas caracteres de espaço, e se qualquer caractere, diferente do espaço for encontrado, a função retorna __False__.

```python
exemplo_0 = ('Neste exemplo demostra a função')
print(exemplo_0.isspace())
```

## ``isnumeric()``

Função ``isnumeric()`` irá verificar se onde ela for aplicada há a existencia de um numero sim ou não, __True__ ou __False__

```python
exemplo_1 = ('Neste exemplo demostra a função')
print(exemplo_1.isnumeric())
```

## ``isalpha()`` 

O método ``isalpha()`` serve para identificar um caractere que não é letra dentro de uma string, ou seja, um caractere inválido. Para demonstrar a existência deste caractere, esse método retornada um __False__

```python
exemplo_2 = ('Neste exemplo demostra a função')
print(exemplo_2.isalpha())
```

## `` isalnum() ``

O método isalnum() retorna True se a string for não-vazia e todos os seus caracteres forem alfanuméricos (ou uma letra ou um número), e caso contrário, retorna False 

```python
exemplo_3 = ('Neste exemplo demostra a função')
print(exemplo_3.isalnum())
```

## ``isupper()``

O método ``isupper()`` retorna “True” se todos os caracteres da string estiverem em maiúsculas, caso contrário, retorna “False”.
Esta função é usada para verificar se o argumento contém algum caractere maiúsculo

```python
exemplo_4 = ('NESTE EXEMPLO')
print(exemplo_4.isupper())
```

## ``islower()``

O método ``islower()`` retorna “True” se todos os caracteres na string estiverem em minúsculas, caso contrário, retorna “False”.
Esta função é usada para verificar se o argumento contém caracteres minúsculos

```python
exemplo_5 = ('exemplo')
print(exemplo_5.islower())
```

## ``istitle()``

O método Python String ``istitle()`` retorna True se a string é uma string com título, caso contrário, retorna False. O que é titulado? A string que tem o primeiro caractere em cada palavra maiúscula e todos os caracteres restantes em letras minúsculas. 

```python
exemplo_6 = ('Exemplo')
print(exemplo_6.istitle())
```

##  ``divmod()``

A função ``divmod()`` da linguagem Python é usada quando queremos efetuar uma __divisão considerando dois números inteiros__ e retornar tanto o __resultado da divisão__ quanto o seu __resto__.

```python
divmod(12,6)
```

## ``dict()``

O ``dict()`` Dicionário em Python é uma coleção de dados sem ordem onde cada elemento possui um par chave/valor. E o que é um par chave/valor? Basicamente é uma forma de se indexar um valor a partir de uma chave.

```python
x = dict(name = "John", age = 36, country = "Norway")
x
```

##  ``dir()``

O ``dir()`` método retorna a lista de atributos válidos do objeto passado. O ``dir()`` método recebe um único parâmetro:

O objeto a inserir pode ser uma tupla vazia/preenchida, lista, conjunto, dicionário etc ou qualquer objeto definido pelo usuário a lista de atributos do objeto passado para o método

```python
number = [12]
# returns valid attributes of the number list 
print(dir(number))
```

## ``.discard()`` 

Função ``.discard()``  descarta ou remove o elemento do conjunto . Este método não retorna nada, nem mesmo nenhum erro se o elemento não estiver presente. Leva um parâmetro que é um elemento a ser removido.

```python
numbers = {2, 3, 4, 5}

# removes 3 and returns the remaining set
numbers.discard(3) 

print(numbers)
```

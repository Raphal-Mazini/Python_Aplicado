# Introdução

Como vimos, classificação é o processo de prever a classe de determinados pontos de dados. Às vezes, as classes são chamadas de destinos / marcadores ou categorias. A modelagem preditiva de classificação é a tarefa de aproximar uma função de mapeamento (f) de variáveis de entrada (X) para variáveis de saída discretas (y).

Por exemplo, a detecção de spam nos provedores de serviços de email pode ser identificada como um problema de classificação. Esta é uma classificação binária, uma vez que existem apenas 2 classes como spam e não spam. Um classificador utiliza alguns dados de treinamento para entender como determinadas variáveis de entrada se relacionam com a classe. Nesse caso, e-mails conhecidos de spam e não-spam devem ser usados como dados de treinamento. 

Quando o classificador é treinado com precisão, ele pode ser usado para detectar um email desconhecido.
A classificação pertence à categoria de aprendizado supervisionado, onde os objetivos também são fornecidos com os dados de entrada. Existem muitas aplicações na classificação em muitos domínios, como aprovação de crédito, diagnóstico médico, marketing de destino etc.

Existem dois tipos de alunos na classificação:

- Lazy Learners

Os Lazy Learners simplesmente armazenam os dados do treinamento e esperam até que os dados do teste apareçam. Quando isso ocorre, a classificação é realizada com base nos dados mais relacionados nos dados de treinamento armazenados. Comparado aos alunos ansiosos, os alunos preguiçosos têm menos tempo de treinamento, mas mais tempo na previsão.
Ex. k-nearest neigbors, case-based reasoning


- Eager Learners

Os Eager Learners constroem um modelo de classificação com base nos dados de treinamento fornecidos antes de receber os dados para classificação. Ele deve poder se comprometer com uma única hipótese que cubra todo o espaço da instância. Devido à construção do modelo, os alunos ansiosos levam muito tempo para treinar e menos tempo para prever.
Ex. Árvore de Decisão, Naive Bayes, Redes Neurais Artificiais

# Naive Bayes


Um classificador Naive Bayes é um modelo probabilístico de aprendizado de máquina usado para tarefas de classificação. O cerne do classificador é baseado no teorema de Bayes.

<br>
\begin{equation*}
P(A | B) = \dfrac{ P(B | A) . P(A) }{ P(B) }
\end{equation*}
<br>


Usando o teorema de Bayes, podemos encontrar a probabilidade de A acontecer, dado que B ocorreu. Aqui, B é a evidência e A é a hipótese. A suposição feita aqui é que os preditores / características são independentes. Essa é a presença de um recurso em particular não afeta o outro. Por isso, é chamado ingênuo.


No final do dia, estamos calculando probabilidades condicionais!.

Para exemplificar:

- Lançamento de Moeda

Quando lançamos uma moeda não viesada (justa), existe uma chance igual de obter cara ou coroa.
Dessa forma, você pode dizer que a probabilidade de sair cara é 50%.

Da mesma forma, qual seria a probabilidade de tirarmos o número 1 em um lançamento de dado?
Novamente, assumindo que o dado é justo, essa probabilidade seria 1/6.

- Baralho de Cartas

Se pegarmos uma carta aleatória do baralho, qual a chance de ser uma rainha, dado que a carta é de espada?

Note que aqui já possuímos a informação de que a carta é de espada. Então diminuímos nosso espaço amostral
Em vez de olharmos para as 52 cartas, vamos observar apenas as 13 cartas de espadas. Então nossa probabildiade
será dada por 1(só existe uma rainha de espada)/ 13 (13 cartas possíveis de espada)

Esse é um exemplo clássico de probabilidade condicional. Então, quando falamos sobre a probabilidade de A
dado B, estamos tentando calcular a probabilidade de A, conhecendo a informação de que B já ocorreu.

Matematicamente, isso pode ser calculado como: P(A|B) = P(A AND B) / P(B)

## Exemplo:

Vamos dar um exemplo para obter uma melhor intuição. Considere o problema de jogar golfe. O conjunto de dados é representado como abaixo.

image.png

Classificamos se o dia é adequado para jogar golfe, dadas as características do dia. As colunas representam esses recursos e as linhas representam entradas individuais. Se pegarmos a primeira linha do conjunto de dados, podemos observar que não é adequado para jogar golfe se a perspectiva estiver chuvosa, a temperatura estiver quente, a umidade estiver alta e não houver vento. Fazemos duas suposições aqui, uma como declarada acima, consideramos que esses preditores são independentes. Ou seja, se a temperatura estiver quente, isso não significa necessariamente que a umidade é alta. Outra suposição feita aqui é que todos os preditores têm um efeito igual no resultado. Ou seja, o dia de vento não tem mais importância na decisão de jogar golfe ou não.



De acordo com este exemplo, o teorema de Bayes pode ser reescrito como:

<br>
<br>
\begin{equation*}
P(y | X) = \dfrac{ P(X | y) . P(y) }{ P(X) }
\end{equation*}
<br>
<br>


A variável y é a variável de classe (jogar golfe), que representa se é adequado jogar golfe ou não dadas as condições. A variável X representa os parâmetros / recursos.
X é dado como:

<br>
\begin{equation*}
X = (x_1, x_2, x_3, ... , x_n)
\end{equation*}
<br>

Aqui x_1, x_2… .x_n representam os recursos, ou seja, eles podem ser mapeados para perspectivas, temperatura, umidade e vento. Substituindo X e expandindo usando a regra da cadeia, obtemos:

<br>
\begin{equation*}
P(y | x_1, x_2, x_3, ... , x_n) = \dfrac{ P(x_1 | y) . P(x_2 | y) . P(x_3 | y) ... P(x_n | y) . P(y) }{ P(x_1).P(x_2).P(x_3) ... P(x_n) }
\end{equation*}
<br>


Agora, você pode obter os valores de cada um observando o conjunto de dados e substituí-los na equação. Para todas as entradas no conjunto de dados, o denominador não muda, permanece estático. Portanto, o denominador pode ser removido e uma proporcionalidade pode ser introduzida.

image.png

No nosso caso, a variável de classe (y) tem apenas dois resultados, sim ou não. Pode haver casos em que a classificação possa ser multivariada. Portanto, precisamos encontrar a classe y com probabilidade máxima.

## Tipos de classificador Naive Bayes:

- Multinomial Naive Bayes:

Isso é usado principalmente para o problema de classificação de documentos, ou seja, se um documento pertence à categoria de esportes, política, tecnologia etc. Os recursos / preditores usados pelo classificador são a frequência das palavras presentes no documento.

- Bernoulli Naive Bayes:

Isso é semelhante aos bayes ingênuos multinomiais, mas os preditores são variáveis booleanas. Os parâmetros que usamos para prever a variável de classe usam apenas valores yes ou no, por exemplo, se uma palavra ocorre no texto ou não.

- Gaussian Naive Bayes:

Quando os preditores assumem um valor contínuo e não são discretos, assumimos que esses valores são amostrados a partir de uma distribuição gaussiana.

## Conclusão

Os algoritmos Naive Bayes são usados principalmente em análise de sentimentos, filtragem de spam, sistemas de recomendação etc. Eles são rápidos e fáceis de implementar, mas sua maior desvantagem é que o requisito dos preditores é independente. Na maioria dos casos da vida real, os preditores são dependentes, o que dificulta o desempenho do classificador.

# Decision Trees (Árvores de Decisão)

Uma árvore tem muitas analogias na vida real e acaba influenciando uma ampla área de aprendizado de máquina, cobrindo tanto a classificação quanto a regressão. Na análise de decisão, uma árvore de decisão pode ser usada para representar visual e explicitamente decisões e tomada de decisão. Como o nome diz, ele usa um modelo de decisão semelhante a uma árvore. Embora seja uma ferramenta comumente usada na mineração de dados para derivar uma estratégia para atingir um objetivo específico, ela também é amplamente usada no aprendizado de máquina, que será o foco principal deste artigo.

## Tipos de Árvore

- **ID3** (Iterative Dichotomiser 3) Usa apenas features categóricas e target categóricos. 


- **C4.5** O algoritmo C4.5 é o sucessor do ID3. Nessa versão, já é possível utilizarmos dados numéricos na nossa classificação.


- **C5.0** Versão melhorada do algoritmo C4.5 que usa menos memória e tende a ser mais preciso.


- **CART** (Classification and Regression Trees) É a a árvore que pode ser utilizada tanto para regressão, quanto para classificação. E, além disso, suporta features numéricas e categóricas.




## Escolha dos nós

Neste procedimento, todos os recursos são considerados e diferentes pontos de divisão são tentados e testados usando uma função de custo. A divisão com o melhor custo (ou o menor custo) é selecionada.

Considere o exemplo anterior de árvore aprendida com o conjunto de dados titanic. Na primeira divisão ou na raiz, todos os atributos / recursos são considerados e os dados de treinamento são divididos em grupos com base nessa divisão. Temos 3 recursos, portanto, temos 3 divisões de candidatos. Agora vamos calcular quanta precisão cada divisão nos custará, usando uma função. A divisão que custa menos é escolhida, que em nosso exemplo é o sexo do passageiro. Esse algoritmo é de natureza recursiva, pois os grupos formados podem ser subdivididos usando a mesma estratégia. Devido a esse procedimento, esse algoritmo também é conhecido como algoritmo ganancioso, pois temos um desejo excessivo de diminuir o custo. Isso torna o nó raiz como melhor preditor / classificador.

### 1. Classificação: 

$$ Gini = \sum_{i} (p_i * (1 - p_i)) $$
$$ Entropy = \sum_{i} p_i * \log_2p_i $$


Uma pontuação de Gini dá uma idéia de quão boa é uma divisão pela mistura das classes de resposta nos grupos criados pela divisão. Aqui, $p_i$ é a proporção de entradas da mesma classe presentes em um grupo específico. Uma pureza perfeita de classe ocorre quando um grupo contém todas as entradas da mesma classe; nesse caso, $p_i$ é 1 ou 0 e G = 0, onde como um nó com uma divisão de 50 a 50 classes em um grupo tem a pior pureza, portanto, para uma classificação binária, terá pk = 0,5 e G = 0,5.

O mesmo reciocínio vale para a Entropia, opu seja, quanto mais baixa a entropia, mais pura e a amostra, sendo que a pureza máxima acontece para valor de entropia igual a zero.

## Quando parar de dividir?

Você pode perguntar quando parar de cultivar uma árvore? Como um problema geralmente possui um grande conjunto de recursos, resulta em grande número de divisões, o que, por sua vez, gera uma árvore enorme. Tais árvores são complexas e podem levar a Overfit. Então, precisamos saber quando parar? Uma maneira de fazer isso é definir um número mínimo de entradas de treinamento para usar em cada folha. Por exemplo, podemos usar um mínimo de 10 passageiros para tomar uma decisão (morrer ou sobreviver) e ignorar qualquer folha que leve menos de 10 passageiros. Outra maneira é definir a profundidade máxima do seu modelo. A profundidade máxima refere-se ao comprimento do caminho mais longo de uma raiz para uma folha.

De maneira geral, quando as divisões realizadas não resultam em incremento de redução de impureza, os nós gerados não são mais considerados.

## Prunning (Poda)

<br>
<img src="https://miro.medium.com/max/547/1*qP8iEcsuVV1sWpFvEHbEDA.png">
<br>

O desempenho de uma árvore pode ser aumentado ainda mais pela poda. Envolve remover as ramificações que fazem uso de recursos de baixa importância. Dessa forma, reduzimos a complexidade da árvore e, assim, aumentamos seu poder preditivo, reduzindo o sobreajuste.

A poda pode começar na raiz (pré-prunning) ou nas folhas (post-prunning).

### Poda Post-prunning (folha):

O método mais simples de poda começa nas folhas e remove cada nó da classe mais popular nessa folha; essa alteração é mantida se não deteriorar a precisão. Também é chamado de remoção de erros reduzida. Exemplos de hiper-parâmetros de poda post-prunning são:

- `min_samples_leaf` : número mínimo de samples por nó folha 

- `max_features`: número de features para o best split

### Poda pré-prunning (raiz):

Métodos de remoção mais sofisticados podem ser usados, como remoção de complexidade de custos, em que um parâmetro de aprendizado (alfa) é usado para avaliar se os nós podem ser removidos com base no tamanho da subárvore. Isso também é conhecido como remoção de link mais fraca.

- `ccp_alpha` : (minimum cost complexity) regulariza a complexidade da árvore pelo nível de impureza das folhas



## Vantagens e Desvantagens

Vantagens do CART (Classification And Regression Trees):

- Simples de entender, interpretar, visualizar.
- As árvores de decisão implicitamente executam seleção de recurso.
- Pode lidar com dados numéricos e categóricos. Também pode lidar com problemas de várias saídas.
- As árvores de decisão requerem relativamente pouco esforço dos usuários para a preparação de dados.
- Os relacionamentos não lineares entre parâmetros não afetam o desempenho da árvore.

Desvantagens do CART (Classification And Regression Trees):

- Os Learners da árvore de decisão podem criar árvores super complexas que não generalizam bem os dados. Isso é chamado de ajuste excessivo ou Overfit.

- As árvores de decisão podem ser instáveis porque pequenas variações nos dados podem resultar na geração de uma árvore completamente diferente. Isso é chamado de variãncia, que precisa ser reduzida por métodos como Bagging e Ensembles.

- Não criam superfícies de decisão suaves. Não retorna resultados contínuos.


Isso é tudo básico, para que você fique a par do aprendizado da árvore de decisão. Uma melhoria sobre o aprendizado da árvore de decisão é feita usando a técnica de aumento. # Iris: Decision Tree Plot

O exemplo ilustra a facilidade conceitual da abordagem da árvore de decisão. No entanto, pode ser dimensionado e aplicado com facilidade a problemas muito maiores. O algoritmo é direto, os recursos e as decisões são claramente definidos em todos os nós internos e os relacionamentos entre os nós são claramente definidos.

Às vezes, é descrito como uma árvore de decisão de aprendizado ou como uma árvore de classificação porque o objetivo é classificar os dados. O exemplo tem dados rotulados ou categóricos. Para variáveis contínuas, árvores de regressão são aplicadas. Nos últimos anos, as técnicas de árvore de decisão foram denominadas Árvores de Classificação e Regressão (CART) à medida que a teoria se expandia, devido à maior disponibilidade de poder computacional.

Crescer uma árvore envolve decidir quais recursos modelar, quais decisões divididas serão aplicadas, usar uma função de custo para avaliar o resultado das divisões e saber quando decidir parar.

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Load data and store it into pandas DataFrame objects
iris = load_iris()
X = pd.DataFrame(iris.data[:, :], columns = iris.feature_names[:])
y = pd.DataFrame(iris.target, columns =["Species"])

# Defining and fitting a DecisionTreeClassifier instance
clf = DecisionTreeClassifier(criterion='gini', min_samples_leaf=10)
clf.fit(X,y)
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```
